```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Declare the constants
Pi = 3.1415926
e = 2.71828
```

```python
#define a function to generate a random number in a particular range.

def randomnumbers(a, b):
    #where a is the lowest limit and b is the highest limit
    return random.uniform(a, b)


#We can define the integrand function, i.e the function we're integrating over
def integrand(x):
    #This is the integral function
    return (e**(-x))/((1+(x-1)**2))
```

```python
def crude_monte_carlo(number_of_samples):
    lower_limit = 0
    upper_limit = 5
    total_of_samples = 0
    
    """This function does the Monte_Carlo_Simulation for f(x) 
    in the range from the lower limit to the upper limit"""
    
    for i in range(number_of_samples):
        x = randomnumbers(lower_limit, upper_limit)
        total_of_samples = total_of_samples + integrand(x)
        
        return (upper_limit - lower_limit) * float(total_of_samples/number_of_samples)
```

```python
number_of_samples = 10000
crude_estimation = crude_monte_carlo(number_of_samples)
#crude_estimation = crude_monte_carlo(MC_samples)

# display results
print(f"Monte Carlo Approximation of f(x): {crude_estimation}")
#print("Monte Carlo Approximation of f(x):" {crude_estimation})
```

### Now we have to define the variance (average of the square of the distances from the mean) of our estimation. 


```python
#The equation above denotes the equation for variance and we'll use this in the simulation
def crude_monte_carlo_variance(num_samples):
    int_max = 5
    
    #for the average of squares 
    running_total = 0
    for i in range(num_samples):
        x = randomnumbers(0, int_max)
        running_total = running_total + integrand(x)**2
    sum_of_squares = running_total* int_max / num_samples
        
    #for the square of average
    running_total = 0
    for i in range(num_samples):
        x = randomnumbers(0, int_max)
        running_total = integrand(x)
    squar_of_average = (int_max*running_total/num_samples)
    
    return sum_of_squares - squar_of_average
    
```

```python
num_samples = 10000
variance = crude_monte_carlo_variance(num_samples)


error = math.sqrt(variance/num_samples)

# display results
print(f"Variance of Approximation: {variance}")
print(f"Error in Approximation: {error}")                          
```
